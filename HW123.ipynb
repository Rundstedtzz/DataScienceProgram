```python
import bootstrapped.bootstrap as bs
import bootstrapped.stats_functions as bs_stats
```

```python
dff2 = np.array(dff['x'])
print(bs.bootstrap(dff2, stat_func=bs_stats.mean))
print(bs.bootstrap(dff2, stat_func=bs_stats.std))
```

```python
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=10, iteration_batch_size=10))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=50, iteration_batch_size=10))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=100, iteration_batch_size=10))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=1000, iteration_batch_size=10))
```

```python
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=10, iteration_batch_size=100))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=50, iteration_batch_size=100))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=100, iteration_batch_size=100))
print(bs.bootstrap(dff2, stat_func=bs_stats.mean, num_iterations=1000, iteration_batch_size=100))
```
